####  Data Preprocessing

```python
del df['customerID']
```

```python
num = num + ['SeniorCitizen ']
```

```python
n_obj = ['gender',
 'SeniorCitizen',
 'Partner',
 'Dependents',
 'PhoneService',
 'MultipleLines',
 'InternetService',
 'OnlineSecurity',
 'OnlineBackup',
 'DeviceProtection',
 'TechSupport',
 'StreamingTV',
 'StreamingMovies',
 'Contract',
 'PaperlessBilling',
 'PaymentMethod']
```

```python
# One hot encoder process
for var in n_obj :
    cat_list = pd.get_dummies(df[var], prefix = var)
    data1 = df.join(cat_list)
    df = data1
```

```python
# Make new dataframe after doing one hot encoder process
data_vars = df.columns.tolist()
col_keep = [i for i in data_vars if i not in n_obj]
df1 = df[col_keep]
```

```python
# Pemisahan feature predictor dan feature target
y = ['Churn']
X = [i for i in df1.columns.values.tolist() if i not in y]
```

```python
# Splitting Data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df1[X], df1[y], test_size=0.1, random_state = 12)
```

```python
print('jumlah data training :',X_train.shape)
print('jumlah data training :',X_test.shape)
```

#### Data Modeling

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
#model 1
model = RandomForestClassifier(n_estimators = 500, max_depth = 10, criterion = 'entropy')

#model 2
model_1 = RandomForestClassifier()
```

```python
#training model 1
model.fit(X_train,y_train)
```

```python
#prediction 1
y_pred = model.predict(X_test)
```

```python
#training model 2
model_1.fit(X_train,y_train)
```

```python
y_pred_1 = model_1.predict(X_test)
```
